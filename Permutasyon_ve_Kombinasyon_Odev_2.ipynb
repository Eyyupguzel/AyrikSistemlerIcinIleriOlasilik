## PERMÜTASYON ve KOMBİNASYON



## Örnek
**(Ahmet, Muhammed, Ali, Deniz, Suzan, Nihat)** 6 kişilik bir öğrenci grubu  TOEFL sınavına girmiştir. 4 öğrenci bu sınavı başarılı bir şekilde geçmiştir. 

A-) 6 Öğrenci arasından 4'ünün başarılı olduğu sınav kaç farklı şekilde sona ermiştir ?

B-) 6 Öğrenci arasından 4'ünün başarılı olduğu kaç farklı altgrup vardır ?


# 1- Permütasyon
Bu problem için, katılan (n) toplam 6 öğrenciden oluşan bir diziden 4 öğrencilik (r) sıralı bir altküme arıyoruz. Başarılı öğrenciler için olası toplam sonuç sayısını bulmak için P(6,4) hesaplamalıyız.

P(6,4) = 6! / (6 - 4)! =720/2= 360 Olası Yarış Sonucu

6 öğrencimizin isimleri  **(Ahmet, Muhammed, Ali, Deniz, Susan, Nihad)**, 4 başarılı öğrenci için 360 olası permütasyonumuz vardır: **('ahmet', 'ali', 'omer', 'deniz'), ('ahmet', 'ali', 'omer', 'yilmaz'),('ahmet', 'ali', 'omer', 'nihat')....vb =360**

# 2- Kombinasyon

Bu problem için, katılan (n) toplam 6 öğrenciden oluşan bir diziden 4  öğrencilik (r) altkümeler arıyoruz. Altkümleri bulmak için C(6,4) hesaplamalıyız.

C(6,4)= 6!/(4! (6-4)!) = 6!/(4!*2!) = 720/48=15

6 öğrencimizin isimleri  **(Ahmet, Muhammed, Ali, Deniz, Susan, Nihad)**, 4 başarılı öğrenci için 15 olası kombinasyonumuz vardır: **('ahmet', 'ali', 'omer', 'deniz'),('ahmet', 'ali', 'omer', 'yilmaz'), ('ahmet', 'ali', 'omer', 'nihat'), ('ahmet', 'ali', 'deniz', 'yilmaz'), ('ahmet', 'ali', 'deniz', 'nihat'), ('ahmet', 'ali', 'yilmaz', 'nihat'), ('ahmet', 'omer', 'deniz', 'yilmaz'), ('ahmet', 'omer', 'deniz', 'nihat'), ('ahmet', 'omer', 'yilmaz', 'nihat')
Comb: ('ahmed', 'deniz', 'yilmaz', 'nihat')
Comb: ('ali', 'omer', 'deniz', 'yilmaz')
Comb: ('ali', 'omer', 'deniz', 'nihat')
Comb: ('ali', 'omer', 'yilmaz', 'nihat')
Comb: ('ali', 'deniz', 'yilmaz', 'nihat')
Comb: ('omer', 'deniz', 'yilmaz', 'nihat') =15**



```python
#Bu, permütasyon ve kombinasyonu bulmak için bir python programıdır.

from math import factorial
from itertools import permutations 
from itertools import combinations 
n=6
r=4

# Permütasyonu bulmak için fonksiyon tanımı
def nPr(n, r):
    npr = factorial(n) // factorial(n - r)
    print(str(n) + "P" + str(r) + ' = ' + str(npr) + "\n")
    perm = permutations(['Ahmet', 'Ali', 'Muhammed', 'Deniz', 'Suzan', 'Nihat'], r) 
    for i in list(perm): 
     print ("perm:",i) 

# Kombinasyonu bulmak için fonksiyon tanımı
def nCr(n, r):
    ncr = factorial(n) // (factorial(r) * factorial(n - r))
    print("\n" + str(n) + "C" + str(r) + ' = ' + str(ncr) + "\n")
    comb = combinations(['Ahmet', 'Ali', 'Muhammed', 'Deniz', 'Suzan', 'Nihat'], r) 
    for i in list(comb): 
      print ("Comb:",i)
    
# Permütasyonu bulmak için işlev çağrısı
nPr(n,r)

# Kombinasyonu bulmak için işlev çağrısı  
nCr(n,r)

```
