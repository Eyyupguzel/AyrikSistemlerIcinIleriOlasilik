```python
from math import factorial
from fractions import Fraction
import numpy as np
from scipy import stats
from scipy.stats import poisson
from scipy.stats import binom
import matplotlib.pyplot as plt


# A) A)	Adil bir zar için X'in olasılık dağılımını oluşturun.
print("A) A)	Adil bir zar için X'in olasılık dağılımını oluşturun:\n")

S={(1,1),(1,2),(1,3),(1,4),(1,5),(1,6),(2,1),(2,2),(2,3),(2,4),(2,5),(2,6),(3,1),(3,2),(3,3),(3,4),(3,5),(3,6),
   (4,1),(4,2),(4,3),(4,4),(4,5),(4,6),(5,1),(5,2),(5,3),(5,4),(5,5),(5,6),(6,1),(6,2),(6,3),(6,4),(6,5),(6,6)}

X2={(1,1)}
X3={(1,2),(2,1)}
X4={(1,3),(2,2),(3,1)}
X5={(1,4),(2,3),(3,2),(4,1)}
X6={(1,5),(5,1),(3,3),(4,2),(2,4)}
X7={(1,6),(6,1),(5,2),(2,5),(4,3),(3,4)}
X8={(5,3),(3,5),(4,4),(6,2),(2,6)}
X9={(4,5),(5,4),(3,6),(6,3)}
X10={(5,5),(6,4),(4,6)}
X11={(6,5),(5,6)}
X12={(6,6)}
PX_2= Fraction(len(X2.intersection(S)),len(S))
print("P(X=2)=",PX_2)
PX_3= Fraction(len(X3.intersection(S)),len(S))
print("P(X=3)=",PX_3)
PX_4= Fraction(len(X4.intersection(S)),len(S))
print("P(X=4)=",PX_4)
PX_5= Fraction(len(X5.intersection(S)),len(S))
print("P(X=5)=",PX_5)
PX_6= Fraction(len(X6.intersection(S)),len(S))
print("P(X=6)=",PX_6)
PX_7= Fraction(len(X7.intersection(S)),len(S))
print("P(X=7)=",PX_7)
PX_8= Fraction(len(X8.intersection(S)),len(S))
print("P(X=8)=",PX_8)
PX_9= Fraction(len(X9.intersection(S)),len(S))
print("P(X=9)=",PX_9)
PX_10= Fraction(len(X10.intersection(S)),len(S))
print("P(X=10)=",PX_10)
PX_11= Fraction(len(X11.intersection(S)),len(S))
print("P(X=11)=",PX_11)
PX_12= Fraction(len(X12.intersection(S)),len(S))
print("P(X=12)=",PX_12)



print("X=x    |  2\t  3\t  4\t  5\t  6\t  7\t  8\t  9\t  10\t  11\t  12\t")
print("---------------------------------------------------------------------------------------------------------------")
print("P(X=x) |",PX_2,"\t" , PX_3,"\t", PX_4,"\t",  PX_5,"\t" , PX_6,"\t",  PX_7,"\t",  PX_8,"\t", PX_9,"\t", PX_10,"\t",  PX_11,"\t", PX_12,"\t\n")

pn=(1/36,2/36,3/36,4/36,5/36,6/36,5/36,4/36,3/36,2/36,1/36)
xn=np.arange(11)
custm = stats.rv_discrete(name='custm', values=(xn, pn))
fig, ax = plt.subplots(1, 1)
ax.plot(xn, custm.pmf(xn), 'ro', ms=12, mec='r')
ax.vlines(xn, 0, custm.pmf(xn), colors='g', lw=10)
plt.title("X rastgele değişkeninin olasılık fonksiyonunun grafiğini: \n")
plt.ylabel("P(x)")
plt.xlabel("x")
plt.show()

#B)	P(X≥9) bulun:
print("B) P(X≥9) bulun:\n")
PX9= (PX_9, PX_10 , PX_11, PX_12)
print("P(X>=9)=",sum(PX9),"-->", float(sum(PX9)),"\n")

#C)	X'in çift bir değer alma olasılığını bulun.
print("C)X'in çift bir değer alma olasılığını bulun:\n")

PXE=(PX_2,PX_4,PX_6,PX_8,PX_10, PX_12)
print("P(X=Çift)=",sum(PXE),"-->", float(sum(PXE)))
```
